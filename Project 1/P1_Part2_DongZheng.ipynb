```python
# Calculate portfolio variance given weight metrix
def portfolio_variance(V, w):
    w_t = w.transpose()
    V_P = w_t @ V @ w
    return V_P[0,0]

# Calculate portfolio return given weight metrix
def portfolio_return(mu, w):
    mu_t = mu.transpose()
    E_P = mu_t @ w
    return E_P[0]
```

```python
#Market portfolio (MKT)
def cal_MKt(Excess_Returns_Market):
    MKt_R=Excess_Returns_Market['Mkt-RF'].describe()['mean']
    MKt_V=(Excess_Returns_Market['Mkt-RF'].describe()['std'])**2
    MKt_std=(Excess_Returns_Market['Mkt-RF'].describe()['std'])
    return (MKt_R,MKt_V,MKt_std)
```

```python
def cal_COV_with_M(Excess_Returns_stocks,real_returns):
    Excess_Returns_Market=Excess_Returns_stocks.copy()
    Excess_Returns_Market['Mkt-RF']=real_returns['Mkt-RF']
    COV_M=Excess_Returns_Market.cov()['Mkt-RF']
    COV_M_Assets= np.array([COV_M.tolist()[:-1]])
    return COV_M_Assets
```

```python
#grand average of beta
def calculate_average_beta(COV_M_Assets,MKt_V):
    COV_M_Assets_list=COV_M_Assets.tolist()
    beta_sum=0
    for cov in COV_M_Assets_list[0]:
        beta_sum+= cov/MKt_V
    grand_average_beta=beta_sum/len(COV_M_Assets_list[0])
    return grand_average_beta
```

```python
def cal_beta(covm,MKt_V):
    a= covm[0].tolist()
    b= [i/MKt_V for i in a]
    return b
```

```python
def cal_CAPM_return(beta_s,MKt_R):
    return np.array([i*MKt_R for i in beta_s])
```

```python
def cal_mu(Excess_Returns_stocks):
    df_mu=Excess_Returns_stocks.describe().loc['mean']
    mu=df_mu.values
    return mu
```

```python
def cal_V(Excess_Returns_stocks):
    #covariance matrix for 48 assets
    Cov_Stocks= Excess_Returns_stocks.cov()
    V= Cov_Stocks.values
    # display(Cov_Stocks)
    return V
```

```python
def cal_CORR(Excess_Returns_stocks):
    Corr_Stocks= Excess_Returns_stocks.corr()
    CORR= Corr_Stocks.values
    # display(Cov_Stocks)
    return CORR
```

```python
def Vshrink_calculator(V,CORR,delta):
    corr_n=np.shape(CORR)[0]
    # print(corr_n)
    aver_cov= (np.sum(CORR)-corr_n)/(corr_n**2-corr_n)
    aver_cov
    VCC= (V/CORR)*aver_cov
    # VCC
    #delta=0.5
    Vshrink= delta*VCC+(1-delta)*V
    return Vshrink
```

```python
#Min-variance portfolio
def weight_GMV(V):
    w_G = np.linalg.inv(V) @ np.ones((48,1))
    w_G /= np.sum(w_G)
    np.round(w_G,5)
    return w_G #transpose?
```

```python
#Tangency portfolio
def weight_TAN(mu,V):
    w_T = np.linalg.inv(V) @ (mu)
    w_T /= np.sum(w_T)
    w_T = np.array([w_T]).T
    return w_T #transpose?
```

```python
#performance metrics
def performance(weight,mu,V):
    R= portfolio_return(mu, weight) #transpose?
    V= portfolio_variance(V, weight)
    STD= math.sqrt(V)
    SR= R/(math.sqrt(V))
    return (R,STD,SR)
```

```python
for d in range(0, 11, 1):
    delta = d /10.0
    MKT_return = []
    MKT_std = []
    MKT_SR = []
    EWP_return = []
    EWP_std = []
    EWP_SR = []
    TAN_return = []
    TAN_std = []
    TAN_SR = []
    TAN_shrink_return = []
    TAN_shrink_std = []
    TAN_shrink_SR = []
    GMV_return = []
    GMV_std = []
    GMV_SR = []
    GMV_shrink_return = []
    GMV_shrink_std = []
    GMV_shrink_SR = []
    #6-fold CV
    for k in range(6):
        #split dataset
        test_data = Excess_Returns_stocks.iloc[60*k:60*(k+1)]
        train_data = Excess_Returns_stocks.drop(test_data.index)
        test_real_returns = real_returns.iloc[60*k:60*(k+1)]
        train_real_returns = real_returns.drop(test_data.index)
        #calculate inputs
        # beta Shrink 
        mkt_train=cal_MKt(train_real_returns)
        covm=cal_COV_with_M(train_data,train_real_returns)
        beta_a=calculate_average_beta(covm,mkt_train[1])
        beta1=cal_beta(covm,mkt_train[1])
        beta_s=[ (0.5*beta_a+0.5*j) for j in beta1]
        #mu CAPM
        mu_CAPM_train = cal_CAPM_return(beta_s,mkt_train[0])
        #V shrink
        V_train = cal_V(train_data)
        CORR_train = cal_CORR(train_data)
        V_s = Vshrink_calculator(V_train,CORR_train,delta)
        #mu
        mu_train=cal_mu(train_data)
        
        #calculate weights of portfolios using inputs on training data
        GMV = weight_GMV(V_train)
        GMV_shrink = weight_GMV(V_s)
        TAN = weight_TAN(mu_train,V_train)
        TAN_shrink = weight_TAN(mu_CAPM_train,V_s)
        
        #calculate and save 3 metrics for each portfolios on testing data
        #Expected return; Standard deviation; Sharpe ratio
        #MKT, EWP, TAN, TAN_shrink, GMV, GMV_shrink
        mu_test=cal_mu(test_data)
        V_test=cal_V(test_data)
        w_E=np.array([[1/48 for n in range(48)]]).transpose()
        
        MKT_return.append(cal_MKt(test_real_returns)[0]) #real returns?
        MKT_std.append(cal_MKt(test_real_returns)[2])
        MKT_SR.append(cal_MKt(test_real_returns)[0]/cal_MKt(test_real_returns)[2])
        EWP_return.append(performance(w_E,mu_test,V_test)[0])
        EWP_std.append(performance(w_E,mu_test,V_test)[1])
        EWP_SR.append(performance(w_E,mu_test,V_test)[2])
        TAN_return.append(performance(TAN,mu_test,V_test)[0])
        TAN_std.append(performance(TAN,mu_test,V_test)[1])
        TAN_SR.append(performance(TAN,mu_test,V_test)[2])
        TAN_shrink_return.append(performance(TAN_shrink,mu_test,V_test)[0])
        TAN_shrink_std.append(performance(TAN_shrink,mu_test,V_test)[1])
        TAN_shrink_SR.append(performance(TAN_shrink,mu_test,V_test)[2])
        GMV_return.append(performance(GMV,mu_test,V_test)[0])
        GMV_std.append(performance(GMV,mu_test,V_test)[1])
        GMV_SR.append(performance(GMV,mu_test,V_test)[2])
        GMV_shrink_return.append(performance(GMV_shrink,mu_test,V_test)[0])
        GMV_shrink_std.append(performance(GMV_shrink,mu_test,V_test)[1])
        GMV_shrink_SR.append(performance(GMV_shrink,mu_test,V_test)[2])
    
    #average
    print(delta, 'MKT_return:',np.mean(MKT_return),'MKT_std',np.mean(MKT_std),'MKT_SR',np.mean(MKT_SR),\
          'EWP_return:',np.mean(EWP_return),'EWP_std',np.mean(EWP_std),'EWP_SR',np.mean(EWP_SR),\
          'TAN_return:',np.mean(TAN_return),'TAN_std',np.mean(TAN_std),'TAN_SR',np.mean(TAN_SR),\
          'TAN_shrink_return:',np.mean(TAN_shrink_return),'TAN_shrink_std',np.mean(TAN_shrink_std),'TAN_shrink_SR',np.mean(TAN_shrink_SR),\
          'GMV_return:',np.mean(GMV_return),'GMV_std',np.mean(GMV_std),'GMV_SR',np.mean(GMV_SR),\
          'GMV_shrink_return:',np.mean(GMV_shrink_return),'GMV_shrink_std',np.mean(GMV_shrink_std),'GMV_shrink_SR',np.mean(GMV_shrink_SR))
    print("---------------------------------------------------------------------------")
  
```
